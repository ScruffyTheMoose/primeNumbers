```python
import sys
sys.path.append('..')

import chebyshevTheta as ct
import sieveOfEratosthenes as soe
import pandas as pd
from math import pow
```

```python
primelist = soe.run(1000)
df = pd.DataFrame(primelist, columns=['primes'])
df
```

```python
thetalist = ct.run(primelist)
df['theta'] = thetalist
df
```

```python
def OTheta(primes: list) -> list:
    
    result = list()
    
    for prime in primes:
        out = prime + pow(prime, (1/2) + 0.000001)
        result.append(out)
        
    return result
```

```python
O_Theta = OTheta(primelist)
O_Theta
```

```python
df['O_Theta'] = O_Theta
df
```

```python
df['raw_error'] = df.apply(lambda row: row['O_Theta'] - row['theta'], axis=1)
df
```

```python
df['%_error'] = df.apply(lambda row: row['raw_error'] / row['O_Theta'], axis=1)
df
```
