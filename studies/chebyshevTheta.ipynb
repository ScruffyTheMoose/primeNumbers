```python
import sys
sys.path.append('..')

import pandas as pd
import chebyshevFunction as pr
import matplotlib.pyplot as plt
```

Retrieving all primes in [2, 1000000000].

```python
df = pd.read_csv('../primes_1000.csv')
```

Building list of all Theta(x) from x = 2 to x = 1000000000

```python
df['theta'] = pr.theta_e(df['primes'].tolist(), len(df))

len(df)
```

Graphically comparing the Theta(x) to x.

```python
plt.figure(figsize=(10, 8), dpi=100)

X = range(len(df))

plt.plot(X, df['theta'], color='r')
plt.plot(X, df['primes'], color='b')

plt.title('Theta(1000000000)')
plt.xlabel('nth Prime')
plt.ylabel('log')

plt.show()
```

Chebyshev's Function is log-linear, and at this scale, it is impossible to visualize the variance of the prime values around Theta(x). We'll need to look at the graph on a smaller scale to actually visualize where x < Theta(x).

The following graph is looking x = 7907, or the first 1000 primes.

```python
plt.figure(figsize=(10, 8), dpi=100)

X = range(1000)

plt.plot(X, df['theta'][:1000], color='r')
plt.plot(X, df['primes'][:1000], color='b')

plt.title('Theta(7907)')
plt.xlabel('nth Prime')
plt.ylabel('log')

plt.show()
```

Now that we can visualize the relationship between x and Theta(x), we can find the specific values where x < Theta(X).

```python
df.loc[df['theta'] > df['primes']]
```

There are no x < Theta(x) because Theta(x) acts as a lower asymptotic bound to x.
